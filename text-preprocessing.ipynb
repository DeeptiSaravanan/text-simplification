#### Paraphrasing

```python
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5")
```

```python
def paraphrasing(phrases):
  para_phrased = []
  for phrase in phrases:
    try:
        para_phrases = parrot.augment(input_phrase=phrase,
                                use_gpu=True,
                                do_diverse=True,
                                adequacy_threshold = 0.80,
                                fluency_threshold = 0.80)
        for para_phrase in para_phrases:
            para_phrased.append(para_phrase)
            break
    except:
        para_phrased.append(phrase)
  return para_phrased
```

```python
kw_model = KeyBERT(model='all-mpnet-base-v2')
```

```python
def extract_keywords(sentence):
  keywords = kw_model.extract_keywords(sentence, keyphrase_ngram_range=(1, 2), stop_words='english', highlight=False, top_n=10)
  keywords_list= list(dict(keywords).keys())
  return keywords_list
```

```python
def synomym_lemma(sentence):

  tokens = nltk.word_tokenize(sentence)
  pos_tags = nltk.pos_tag(tokens)

  simplified_sentence = []
  keywords_list = extract_keywords(sentence)
  
  for word, pos in pos_tags:
      if(word in phrase.split() for phrase in keywords_list):
          if pos.startswith('NN'):
              synsets = wordnet.synsets(word, pos=wordnet.NOUN)
              if synsets:
                  lemma = synsets[0].lemmas()[0]
                  simplified_word = lemma.name().replace('_', ' ').lower()
                  simplified_sentence.append(simplified_word)
              else:
                  simplified_sentence.append(word)
          elif pos.startswith('VB'):
              synsets = wordnet.synsets(word, pos=wordnet.VERB)
              if synsets:
                  lemma = synsets[0].lemmas()[0]
                  simplified_word = lemma.name().replace('_', ' ').lower()
                  simplified_sentence.append(simplified_word)
              else:
                  simplified_sentence.append(word)
          else:
              simplified_sentence.append(word)

  simplified_sentence = ' '.join(simplified_sentence)

  return simplified_sentence
```

```python
def generate_questions(simplified_sentence):
  keywords_list = extract_keywords(simplified_sentence)
  questions = []
  print(keywords_list)
  for keyword in keywords_list:
    questions.append(f"What is {keyword}?")
    questions.append(f"What does {keyword} do?")
    questions.append(f"How does {keyword} work?")
  return questions
```

```python
def get_ner(simplified_sentence):
  nlp = spacy.load('en_core_web_sm')
  doc = nlp(simplified_sentence)
  ner = []

  for ent in doc.ents:
    ner.append(ent.text)
  return ner
```

```python
def mrc(simplified_sentence, questions, ner):
  mrc_model = pipeline('question-answering', model='distilbert-base-cased-distilled-squad')

  answers = []
  for question in questions:
    answer = mrc_model(question=question, context=simplified_sentence)
    if(answer['answer'] not in ner and answer['answer'] not in answers):
      answers.append(answer['answer'])
  return answers
```

```python
def get_verbs(keytokens):
  verbs = []
  for input_sentence in keytokens:
    words = word_tokenize(input_sentence)
    tagged_words = pos_tag(words)
    input_verbs = [word for word, tag in tagged_words if tag.startswith('VB')]
    verbs.extend(input_verbs)
  verbs = sorted(list(set(verbs)), key=len, reverse=True)
  return verbs
```

```python
def gpt2(text, verbs, mask_token="[MASK]"):
  model_name = "gpt2"
  tokenizer = transformers.GPT2Tokenizer.from_pretrained(model_name)
  config = transformers.GPT2Config.from_pretrained(model_name)
  model = transformers.GPT2LMHeadModel.from_pretrained(model_name, config=config)

  tokens = tokenizer.tokenize(text)
  mask_positions = []
  for word in tokens:
    if(word in verbs):
      mask_positions.append(tokens.index(word))
  for i in mask_positions:
      tokens[i] = mask_token
  masked_text = tokenizer.convert_tokens_to_string(tokens)
  input_ids = tokenizer.encode(masked_text, return_tensors="pt")
  with torch.no_grad():
      outputs = model(input_ids)
      predictions = outputs[0]
  for i in mask_positions:
      predicted_token_id = torch.argmax(predictions[0, i]).item()
      predicted_token = tokenizer.convert_ids_to_tokens([predicted_token_id])[0]
      tokens[i] = predicted_token
  filled_text = tokenizer.convert_tokens_to_string(tokens)
  return filled_text  
```

```python
def two_way_translation(sentence):

  blob = TextBlob(sentence)
  translation = blob.translate(from_lang='en', to='fr')
  out = translation.translate(from_lang='fr', to='en')
  return out
```

```python
def main(data):
  para_phrased = paraphrasing(data)
  print(para_phrased)
  print('----------------------')
  outputs = []
  for sentence in para_phrased:
    #print("Running Lemmatization and Synonyms Replacement")
    simplified_sentence = synomym_lemma(sentence[0])
    print(simplified_sentence)
    print('----------------------')
    #print("Generating questions")
    questions = generate_questions(simplified_sentence)
    print(questions)
    print('----------------------')
    ner = get_ner(simplified_sentence)
    #print("Generating Answers")
    answers = mrc(simplified_sentence, questions, ner)
    print(answers)
    print('----------------------')
    verbs = get_verbs(answers)
    filled_text = gpt2(simplified_sentence,verbs)
    print(filled_text)
    print('----------------------')
    translation_out = two_way_translation(filled_text)
    final_output=''
    for i in range(len(translation_out.tokens)):
      if(i == len(translation_out.tokens)-1):
        final_output = final_output + translation_out.tokens[i]
      else:
        final_output = final_output + translation_out.tokens[i]
        final_output+=" "
    outputs.append(final_output)
    print(outputs)
  return outputs
```

#### Depth of dependency tree

```python
import spacy

# Load the language model
nlp = spacy.load("en_core_web_sm")

# Define the sentence to be analyzed
sentences = dataset['test']['long'][0:10]

for sentence in sentences:
  # Process the sentence using the language model
  doc = nlp(sentence)

  # Find the root of the dependency tree
  root = None
  for token in doc:
      if token.head == token:
          root = token
          break

  # Calculate the depth of the tree recursively
  def get_depth(token):
      if token == root:
          return 0
      else:
          return 1 + get_depth(token.head)

  depth = max([get_depth(token) for token in doc])

  print(f"The depth of the dependency tree is {depth}")
```

#### Number of Clauses (Complex should have more)

```python
import nltk

# Download the necessary NLTK resources
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

sentences = dataset['test']['long'][0:10]

for sentence in sentences:

  print(sentence)

  # Tokenize the sentence into clauses
  clauses = nltk.sent_tokenize(sentence)

  # Identify the dependent clauses using part-of-speech tagging
  dependent_clauses = []
  for clause in clauses:
      words = nltk.word_tokenize(clause)
      tags = nltk.pos_tag(words)
      for i in range(len(tags)):
          if tags[i][1] == 'IN':  # IN tag indicates a subordinating conjunction
              dependent_clause = ' '.join(words[i+1:])
              dependent_clauses.append(dependent_clause)

  # Print the result
  if len(dependent_clauses) > 0:
      #print("The dependent clause(s) in the sentence are:")
      print(len(dependent_clauses))
      for clause in dependent_clauses:
          print(clause)
  else:
      #print("The sentence does not contain any dependent clauses.")
      print('0')
  print('--------------------------------')
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')


def clause_dependency_ratio(complex_sentence, simple_sentence):


 comp_clauses = nltk.sent_tokenize(complex_sentence)
 simple_clauses = nltk.sent_tokenize(simple_sentence)


 comp_dependent_clauses = []
 simple_dependent_clauses = []
 for clause in comp_clauses:
     words = nltk.word_tokenize(clause)
     tags = nltk.pos_tag(words)
     for i in range(len(tags)):
         if tags[i][1] == 'IN':  # IN tag indicates a subordinating conjunction
             dependent_clause = ' '.join(words[i+1:])
             comp_dependent_clauses.append(dependent_clause)


 for clause in simple_clauses:
     words = nltk.word_tokenize(clause)
     tags = nltk.pos_tag(words)
     for i in range(len(tags)):
         if tags[i][1] == 'IN':  # IN tag indicates a subordinating conjunction
             dependent_clause = ' '.join(words[i+1:])
             simple_dependent_clauses.append(dependent_clause)


 return int(len(comp_dependent_clauses)/len(simple_dependent_clauses))
```

#### Number of distinct POS Tags (Expecting more in complex sentence)

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')


def pos_compression_ratio(complex_sentence, simple_sentence):


 comp_words = nltk.word_tokenize(complex_sentence)
 simple_words = nltk.word_tokenize(simple_sentence)
 comp_tags = nltk.pos_tag(comp_words)
 comp_distinct_tags = set(tag[1] for tag in comp_tags)
 simple_tags = nltk.pos_tag(simple_words)
 simple_distinct_tags = set(tag[1] for tag in simple_tags)


 return int(len(comp_distinct_tags)/len(simple_distinct_tags))
```

```python
import nltk

# Download the necessary NLTK resources
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

sentences = dataset['test']['long'][0:10]

for sentence in sentences:

  # Tokenize the sentence into words
  words = nltk.word_tokenize(sentence)

  # Tag the words with their POS
  tags = nltk.pos_tag(words)

  # Count the number of distinct POS tags
  distinct_tags = set(tag[1] for tag in tags)
  num_distinct_tags = len(distinct_tags)

  # Print the result
  print("The sentence contains", num_distinct_tags, "distinct POS tags.")
  print(distinct_tags)
```
